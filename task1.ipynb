# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
The objective of this task is to explore the Iris dataset using Python and create visualizations to understand the relationships between different flower species.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries Imported Successfully!")
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Scatter Plot: Sepal Length vs Sepal Width

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    palette="Set1",
    s=80
)

plt.title("Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.show()
```

### Observation

- The scatter plot shows the relationship between sepal length and sepal width.
- Setosa flowers are clearly separated from the other two species.
- Versicolor and Virginica have some overlapping values.

```python
# Histogram of all numerical features

df.drop("species", axis=1).hist(
    figsize=(10,6),
    bins=20,
    edgecolor="black"
)

plt.suptitle("Distribution of All Features")

plt.show()
```

### Observation

- The histograms show how the values are distributed.
- Petal length and petal width vary significantly among the flower species.
- Most values are concentrated around the center of each distribution.

```python
# Box plots for each feature

fig, axes = plt.subplots(2,2, figsize=(12,8))

features = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width"
]

for i, feature in enumerate(features):
    sns.boxplot(
        data=df,
        x="species",
        y=feature,
        ax=axes[i//2, i%2]
    )

    axes[i//2, i%2].set_title(feature)

plt.tight_layout()

plt.show()
```

### Observation

- Setosa has the smallest petal measurements.
- Virginica has the largest petal measurements.
- Only a few outliers are present in the dataset.
